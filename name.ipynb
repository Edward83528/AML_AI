```python
import json; #爬網站資料格式需要
import requests; #爬網站資料需要
```

```python
with open('AML_NER.json','r',encoding="utf-8") as load_f:
    load_list = json.load(load_f)


dict_name = dict()
print(len(load_list))
for i in range(0,len(load_list)-1):
    for name in load_list[i]['person'].split(','):
        if (len(name)>1) and (load_list[i]['name'].find(name) != -1):
            dict_name[name] = 1
        else:
            dict_name[name] = 0
print('done')    
```

```python
f = open('train_name_classification.csv','w',encoding='utf-8')
f.write('name,label\n')
for key, value in dict_name.items():
    f.write(str(key) + ',' + str(value) + '\n')
```
