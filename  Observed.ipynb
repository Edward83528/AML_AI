```python
import json; #爬網站資料格式需要
import requests; #爬網站資料需要
```

```python
with open('AML_NER.json','r',encoding="utf-8") as load_f:
    load_list = json.load(load_f)

#print(type(load_list))
count=0
for i in range(0,len(load_list)-1):
    #if len(load_list[i]['context'])>510 and load_list[i]['name'] != '[]': # 120項符合
    if load_list[i]['name'] != '[]':
        print('context:',load_list[i]['context'])
        print('title:',load_list[i]['title'])
        print('name:',load_list[i]['name'])
        count+=1
        print('---'*50)

print("count:",count)    
```
